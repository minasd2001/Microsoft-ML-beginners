```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model, model_selection
```

```python
from sklearn.datasets import load_linnerud

linnerud = load_linnerud()

print("Exercise feature names:", linnerud.feature_names)
print("target names: ", linnerud.target_names)
```

```python
X = pd.DataFrame(data=linnerud.data, columns=linnerud.feature_names)
```

```python
y = pd.DataFrame(data = linnerud.target , columns = linnerud.target_names)
```

```python
print(X.head())
```

# Plotting each targets with each feature

```python
fig, axes = plt.subplots(3, 3, figsize=(12, 12))

for i, feature in enumerate(X.columns):
    for j, target in enumerate(y.columns):
        axes[i, j].scatter(X[feature], y[target], color="blue", alpha=0.7)
        axes[i, j].set_xlabel(feature)
        axes[i, j].set_ylabel(target)
        axes[i, j].set_title(f"{feature} vs {target}")

plt.tight_layout()
plt.show()

```
